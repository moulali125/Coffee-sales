```python
df['date']=pd.to_datetime(df['date'])
df['datetime']=pd.to_datetime(df['datetime'])
#column for month,Weekdays and hours
df['month']=df['date'].dt.strftime('%Y-%m')
df['day']=df['date'].dt.strftime('%w')
df['hour']=df['datetime'].dt.strftime('%H')
```

```python
rd=df.groupby(['coffee_name']).sum(['money']).reset_index().sort_values(by='money',ascending=False)
```

```python
rd
```

```python
plt.figure(figsize=(10,4))
ax=sns.barplot(data=rd, x='money',y='coffee_name')
ax.bar_label(ax.containers[0],fontsize=8)
plt.xlabel('Revenue')
```

```python
from operator import index
monthly_sales=df.groupby(['coffee_name','month']).count()['date'].reset_index().rename(columns={'date':'count'}).pivot(index='month',columns='coffee_name',values='count').reset_index()
monthly_sales
```

```python
monthly_sales.describe().T.loc[:,['min','max']]
```

```python
plt.figure(figsize=(12,6),facecolor='blue') # Use facecolor instead of color
sns.lineplot(data=monthly_sales) # Pass the DataFrame to data
plt.legend(loc='upper right')
plt.xticks(range(len(monthly_sales['month'])),monthly_sales['month'],size='small')
plt.xlabel('Month')
```

```python
weekday_sales=df.groupby(['day']).count()['date'].reset_index().rename(columns={'date':'count'})
weekday_sales
```

```python
plt.figure(figsize=(10,4))
sns.barplot(data=weekday_sales,x='day',y='count')
plt.xticks(range(len(weekday_sales['day'])),['Sun','Mon','Tue','Wed','Thu','Fri','Sat'])
plt.xlabel('Weekday')
```

```python
daily_sales=df.groupby(['coffee_name','date']).count()['datetime'].reset_index().reset_index().rename(columns={'datetime':'count'}).pivot(index='date',columns='coffee_name',values='count').reset_index().fillna(0)
daily_sales

```

```python
daily_sales.iloc[:,1:].describe().T.loc[:,['min','max']]
```

```python
hourly_sales=df.groupby(['hour']).count()['date'].reset_index().rename(columns={'date':'count'})
hourly_sales
```

```python
sns.barplot(data=hourly_sales,x='hour',y='count')
```

```python
hourly_sales_by_coffee=df.groupby(['hour','coffee_name']).count()['date'].reset_index().rename(columns={'date':'count'}).pivot(index='hour',columns='coffee_name',values='count').fillna(0).reset_index()
hourly_sales_by_coffee
```

```python
fig,axs=plt.subplots(nrows=2, ncols=4, figsize=(15,5)) # Create a 3x3 grid of subplots

axs=axs.flatten() # Flatten the 2D array of Axes objects into a 1D array
for i, column in enumerate(hourly_sales_by_coffee.columns[1:]): # Use enumerate to get both index and value
  axs[i].bar(hourly_sales_by_coffee['hour'],hourly_sales_by_coffee[column])
  axs[i].set_title(f'Coffee: {column}')
  axs[i].set_xlabel('Hour')
  axs[i].set_ylabel('Count')
plt.tight_layout()

plt.show()
```
